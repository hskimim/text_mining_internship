```python
from selenium import webdriver
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup
import time
from scrapy.http import TextResponse
```

```python
%%time
title_ls = []
clicked_ls = []
time_ls = []
link_ls = []

for page in range(1,500+1):
    if page % 100 ==0 : print('{} page is over.'.format(page))
    url = 'https://coinkorea.info/index.php?mid=community&page={}'.format(page)
    req = requests.get(url)
    response = TextResponse(req.url, body=req.text, encoding='utf-8')
    for content in range(6,25+1):
        try :
            try : 
                title_ls.append(response.xpath('//*[@id="main"]/div/div/div/div[1]/div/div/table/tbody/tr[{}]/td[2]/a/span[1]/text()'.format(content)).extract()[0])
            except : 
                title_ls.append(response.xpath('//*[@id="main"]/div/div/div/div[1]/div/div/table/tbody/tr[{}]/td[2]/a/span[1]/span/text()'.format(content)).extract()[0])
            clicked_ls.append(response.xpath('//*[@id="main"]/div/div/div/div[1]/div/div/table/tbody/tr[{}]/td[6]/text()'.format(content)).extract()[0])
            time_ls.append(response.xpath('//*[@id="main"]/div/div/div/div[1]/div/div/table/tbody/tr[{}]/td[5]/text()'.format(content)).extract()[0])
            link_ls.append(response.xpath('///*[@id="main"]/div/div/div/div[1]/div/div/table/tbody/tr[{}]/td[2]/a/@href'.format(content)).extract()[0])
        except : print(page,content)
```

```python
df = pd.DataFrame()

df['title'] = title_ls
df['time'] = time_ls
df['click'] = clicked_ls
df['link'] = link_ls
```
