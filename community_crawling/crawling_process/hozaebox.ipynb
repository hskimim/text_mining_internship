```python
from selenium import webdriver
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup
import time
from scrapy.http import TextResponse
```

```python
%%time
title_ls = []
clicked_ls = []
time_ls = []
link_ls = []

for page in range(1,147+1):
    if page % 100 ==0 : print('{} page is over.'.format(page))
    url = 'http://hozaebox.com/bbs/board.php?bo_table=freeboard&page={}'.format(1)
    req = requests.get(url)
    response = TextResponse(req.url, body=req.text, encoding='utf-8')
    for content in range(1,15+1):
        try : 
            testing_ls = response.xpath('//*[@id="fboardlist"]/table/tbody/tr[{}]/td[2]/a/text()'.format(content)).extract()[1]
            testing_ls = testing_ls[1:].replace(' ','')
            title_ls.append(testing_ls)
            clicked_ls.append(response.xpath('//*[@id="fboardlist"]/table/tbody/tr[{}]/td[5]/text()'.format(content)).extract()[0])
            time_ls.append(response.xpath('//*[@id="fboardlist"]/table/tbody/tr[{}]/td[4]/text()'.format(content)).extract()[0])
            link_ls.append(response.xpath('//*[@id="fboardlist"]/table/tbody/tr[{}]/td[2]/a/@href'.format(content)).extract()[0])
        except : print('pass')
```

```python
df3 = pd.DataFrame()

df3['title'] = title_ls
df3['time'] = time_ls
df3['click'] = clicked_ls
df3['link'] = link_ls
df3.tail()
```

```python
%%time
content_ls = []
for idx in range(len(df3)):
    if idx % 10 ==0 : print(idx)
    req = requests.get(df3['link'].values[idx])
    response = TextResponse(req.url, body=req.text, encoding='utf-8')
    process_ls = ','.join(response.xpath('//*[@id="bo_v_con"]/p/text()').extract()).replace(',','')
    content_ls.append(process_ls)
```

```python
df4 = pd.DataFrame()

df4['title'] = title_ls
df4['time'] = time_ls
df4['click'] = clicked_ls
df4['content'] = content_ls
```

```python
df4.to_csv('coinpan_data.csv',index=False)
```
